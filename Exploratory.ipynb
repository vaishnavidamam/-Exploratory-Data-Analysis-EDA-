```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')  # Comes preloaded in seaborn
titanic.head()
```

```python
titanic.describe(include='all')
```

```python
titanic.info()
```

```python
titanic['class'].value_counts() 
```

```python
sns.pairplot(titanic.dropna(), hue='survived')  # Drop NA to avoid errors
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
titanic['age'].hist(bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=titanic)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs Fare colored by Survival')
plt.show()
```
